## Missing Value Analysis

```python
#set Time as Index column
raw_dataset = raw_dataset.set_index("Time")
raw_dataset.head()
```

```python
#calculate the percentage of NA values in our data
raw_dataset.isna().sum().sum()/raw_dataset.size #sum() 2 times to have the total number of NA value
#raw_dataset.isna().sum(axis = 1) showing the number of NA per row
```

```python
#Calculate percentage of NA values per column and plot them on histogram
perc_na = raw_dataset.isna().sum()/len(raw_dataset)
sns.histplot(perc_na,bins=40)
#=====> observation: there are some 80% missing value, hence these could be missing or malfunction devices. 
```

```python
perc_na.sort_values(ascending=False)
```

```python
#Now we plot the most 5 missing values Meters and see what's going on
low_value_meters = perc_na.sort_values(ascending=False).index[0:5]
low_value_meters

fig,ax = plt.subplots(nrows=5,figsize = (20,15))
for i, column in enumerate(low_value_meters):
    plot1 = sns.lineplot(raw_dataset[column],ax=ax[i])
    plot1.set_xlim(raw_dataset.index.min(), raw_dataset.index.max()) #set limit of x axis, since Time is the Index column -> plot based on Min time and Max time series.

#Seemed like Meter 38 is malfunction around 2022-06
```

```python
#Now we plot the lease 5 missing values Meters and see what's going on
high_value_meters = perc_na.sort_values(ascending=True).index[0:5]
high_value_meters

fig,ax = plt.subplots(nrows=5,figsize = (20,15))
for i, column in enumerate(high_value_meters):
    plot1 = sns.lineplot(raw_dataset[column],ax=ax[i])
    plot1.set_xlim(raw_dataset.index.min(), raw_dataset.index.max()) #set limit of x axis, since Time is the Index column -> plot based on Min time and Max time series.

#Seemed like missing values between the 202306 and 202307
#Meter140 has weird record
```

```python
number_of_meters = raw_dataset.shape[1]
```

```python
nan_by_date = raw_dataset.isna().sum(axis = 1)
nan_by_date = nan_by_date.rename("NA_count").reset_index()
nan_by_date

percent_nan_by_date = nan_by_date.copy()
percent_nan_by_date.NA_count = 100*percent_nan_by_date.NA_count / number_of_meters
fig,ax = plt.subplots(ncols=2,figsize = (20,5))
#plot by each day how many NA values. We can see from 1st graph that timerange 202307-202309 NA_count is significantly high.
sns.lineplot(nan_by_date, x= "Time", y="NA_count", ax =ax[0])
#draw 2nd graph to analyse the downtime 202307-202309
sns.lineplot(percent_nan_by_date[-10000:],x= "Time", y="NA_count", ax =ax[1]) #last 10000 elements
ax[1].set_ylabel("% devices down")
```

```python
# Remove duplicates in the correlation
il = np.tril_indices(number_of_meters) #only
il
triang_corr = corr_mat.copy().to_numpy()
triang_corr[il] = np.nan
triang_corr = pd.DataFrame(triang_corr, columns=corr_mat.columns, index = corr_mat.index)
sns.heatmap(triang_corr)
```

```python
#Extract devices that are highly correlate to each other or negetively correlate
xx = triang_corr[(triang_corr >0.5)&(triang_corr <1) | (triang_corr <=-0.5)]
#Print out the pairs
for i in xx: #row
    for j, cell in enumerate(xx[i]): #col
        if np.isnan(cell) == False:
            print(i, xx.columns[j], cell)
```

```python
#Different ways to show the pairs without the nested for loop

# stack turns the matrix into long format (row, col, value)
pairs = triang_corr.stack().reset_index()
pairs
pairs.columns = ["Meter1", "Meter2", "Correlation"]
strong_pairs = pairs[(pairs["Correlation"].abs() >= 0.5) & (pairs["Correlation"] < 1)]
strong_pairs
```
